```python
import pandas as pd

df = pd.read_csv('18.csv')

df['creationdate'] = pd.to_datetime(df['creationdate'])

questions_before_2014 = df[df['creationdate'] < '2014-01-01']
print(" 2014-yildan oldin yaratilgan savollar:\n", questions_before_2014)

high_score_questions = df[df['score'] > 50]
print("\n Skori 50 dan katta savollar:\n", high_score_questions)

medium_score_questions = df[(df['score'] >= 50) & (df['score'] <= 100)]
print("\n Skori 50 va 100 orasida bo‘lgan savollar:\n", medium_score_questions)

scott_questions = df[df['ans_name'] == 'Scott Boston']
print("\n Scott Boston javob bergan savollar:\n", scott_questions)

users = ['User1', 'User2', 'User3', 'User4', 'User5']
questions_by_users = df[df['ans_name'].isin(users)]
print("\n Ushbu 5 foydalanuvchi javob bergan savollar:\n", questions_by_users)

unutbu_questions = df[
    (df['creationdate'] >= '2014-03-01') &
    (df['creationdate'] <= '2014-10-31') &
    (df['ans_name'] == 'Unutbu') &
    (df['score'] < 5)
]
print("\n Unutbu javob bergan va 2014 mart-oktabr oralig‘ida yaratilgan, skori <5 bo‘lgan savollar:\n", unutbu_questions)

filtered_questions = df[(df['score'].between(5, 10)) | (df['viewcount'] > 10000)]
print("\n Skori 5-10 oralig‘ida yoki ko‘rish soni 10,000 dan katta savollar:\n", filtered_questions)

not_scott_questions = df[df['ans_name'] != 'Scott Boston']
print("\n Scott Boston javob bermagan savollar:\n", not_scott_questions)
```
